```python
import pandas as pd
import numpy as np
import random
```

```python
num_rows = 300

data = {
    "CustomerID": [f"CUST{i:04}" for i in range(1, num_rows + 1)],
    "Age": np.random.randint(21, 65, size=num_rows),
    "Gender": np.random.choice(['Male', 'Female'], size=num_rows),
    "Income": np.random.randint(20000, 150000, size=num_rows),
    "LoanAmount": np.random.randint(10000, 100000, size=num_rows),
    "LoanTenure": np.random.randint(6, 60, size=num_rows),  # in months
    "CreditScore": np.random.randint(300, 850, size=num_rows),
    "NumOfLatePayments": np.random.randint(0, 6, size=num_rows),
    "Defaulted": np.random.choice(['Yes', 'No'], size=num_rows)
}

df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv('/content/credit_data.csv', index=False)
```

```python
df = pd.read_csv('/content/credit_data.csv')
```

```python
def assign_risk(row):
    score = 0

    if row['CreditScore'] < 580:
        score += 2
    elif row['CreditScore'] < 670:
        score += 1

    if row['NumOfLatePayments'] >= 3:
        score += 2
    elif row['NumOfLatePayments'] >= 1:
        score += 1

    if row['Income'] < 30000:
        score += 1

    if row['LoanAmount'] > 0.4 * row['Income']:
        score += 1

    if score >= 4:
        return 'High'
    elif score >= 2:
        return 'Medium'
    else:
        return 'Low'

df['RiskLevel'] = df.apply(assign_risk, axis=1)
```

```python
df.to_csv('/content/cleaned_credit_data.csv', index=False)
```
